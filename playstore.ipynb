# Data Visualization 

```python
Type_count = apps['Category'].value_counts()
plt.pie(Type_count.values , labels = apps['Category'].unique() , startangle=90 , autopct = '%1.1f%%' , shadow = True , radius = 10 , rotatelabels = 90)
plt.figure(figsize = (12,7) )
plt.show()
```

```python
Type_count = apps['Category'].value_counts()
plt.figure(figsize = (12,7) )
sns.barplot (Type_count.index , Type_count.values , alpha = 0.9)
plt.title ('Count of App installed by Category ')
plt.xlabel ('Category' , fontsize = 12)

plt.ylabel ('Number of Apps' , fontsize = 12)
plt.xticks(rotation=90)
plt.show()
```
